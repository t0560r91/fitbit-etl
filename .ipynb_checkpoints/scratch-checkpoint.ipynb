```python
import json
import requests
```

```python
def request_heart_data(token: dict, date: str, level='1min') -> dict:
    """Requests intraday heart data.
    """
    user_id = token['user_id']
    access_token = token['access_token']
    refresh_token = token['refresh_token']
    token_type = token['token_type']

    endpoint = f'https://api.fitbit.com/1.2/user/{user_id}/activities/heart/date/{date}/1d/{level}/time/00:00/23:59.json'
    headers = {
        'content-type' : 'application/json', 
        'authorization' : token_type + " " + access_token
        }

    heart_data = json.loads(requests.get(endpoint, headers=headers).text)
    if 'activities-heart-intraday' in heart_data.keys(): 
        return heart_data # pass
    else:
        if 'success' in heart_data.keys():
            raise ValueError('Expired Token')
        else:
            raise ValueError('Invalid Date Input')
            # 150 requests/hour limit? 
```

```python
pd.options.display.max_rows = None

def request_multiple_heart_data(token, start_date, end_date):
    user_id = token['user_id']
    result_df = pd.DataFrame()
    dates = pd.date_range(start_date, end_date)
    for d in dates:
        data = request_heart_data(token, d.strftime('%Y-%m-%d'))
        date = data['activities-heart'][0]['dateTime']
        df = pd.DataFrame(data['activities-heart-intraday']['dataset'])
        df['date'] = [date] * df.shape[0]
        df['userid'] = [user_id] * df.shape[0]
        df_t = pd.DataFrame([df['date'], df['time'], df['value'], df['userid']]).transpose()
        result_df = pd.concat([result_df, df_t], axis=0)
        result_df.reset_index(drop=True, inplace=True)

        
    return result_df
```

```python
data = request_multiple_heart_data(token, '2019-08-01', '2019-08-03')
```
